```python
pixel_width = 10
pixel_height = 10
plt.figure(figsize=(pixel_width, pixel_height))
# 将均值文件读入blob中
mean_blob = caffe.proto.caffe_pb2.BlobProto()
mean_blob.ParseFromString(open(MEAN_FILE, 'rb').read())

# 将均值blob转为numpy.array
mean_npy = caffe.io.blobproto_to_array(mean_blob)
# 均值参数是多维数组，为了方便输出，转为单列数组
print mean_npy[0].shape
plt.imshow(mean_npy[0].transpose(1,2,0))
plt.axis('on')
plt.show()
# plt.show()
# mean_npy.shape = (-1, 1)
# for m in mean_npy:
#     # 写参数
#     mf.write('%ff, ' % m)

# mf.close
```
